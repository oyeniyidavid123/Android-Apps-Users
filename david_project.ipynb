```python
data_df=data_df.rename(columns={"App Name":"App_Name"})
```

```python
data_df.head(50)
```

```python
#selecting columns that are relevating in the dataset
data_df=data_df[["App_Name","Category","Rating","Reviews","Installs","Size","Price","Content Rating"]]
```

```python
#drop missing columns and rows
data_df.dropna()
```

```python
#data cleaning
data_df=data_df[~data_df.App_Name.str.contains("?", regex=False)]
```

```python
data_df.head(50)
```

```python
data_df.dtypes
```

```python
data_df.count()
```

```python
#data cleaning
data_df=data_df[~data_df.Price.str.contains("M", "+", na=False,regex=False)]
```

```python
#clean price
data_df["Price"] = data_df["Price"].str.replace("$","")
data_df = data_df[data_df["Price"] != 'Varies with device']
```

```python
#cleaning Installs
data_df = data_df[data_df["Installs"] != 'EDUCATION']
data_df = data_df[data_df["Installs"] != ' Xmax X']
data_df["Installs"] = data_df['Installs'].str.split("+").str[0]
data_df["Installs"] = data_df["Installs"].replace(',', "", regex=True)
data_df["Installs"] = data_df["Installs"].astype(float)
```

```python
#cleaning Rating
data_df = data_df[data_df["Rating"] != 'NEWS_AND_MAGAZINES']
data_df = data_df[data_df["Rating"] != 'GAME_STRATEGY']
data_df["Rating"] = data_df["Rating"].astype(float)
```

```python
data_df.head(50)
```

```python
#convert object to float
#data_df.loc[:, "Price"] = data_df["Price"].astype(float)

```

```python
data_df.dtypes
```

```python
data_df["Category"]=data_df["Category"].replace('_', ' ', regex=True)
clean_df = data_df.replace(
    {"GAME PUZZLE": "GAME",
"GAME CASUAL": "GAME",
"GAME ARCADE": "GAME",
"GAME ACTION": "GAME",
"GAME SIMULATION": "GAME",
"GAME CARD": "GAME",
"GAME EDUCATIONAL": "GAME",
"GAME ROLE PLAYING": "GAME",
"GAME STRATEGY": "GAME",
"GAME RACING": "GAME",
"GAME ADVENTURE": "GAME",
"GAME SPORTS": "GAME",
"GAME WORD": "GAME",
"GAME BOARD": "GAME",
"GAME CASINO": "GAME",
"GAME TRIVIA": "GAME",
"GAME MUSIC": "GAME",
})
clean_df["Category"].value_counts()


```

```python
cleaned_df2 = clean_df.groupby(["Category"])
```

```python
Total_Review = cleaned_df2["Reviews"].sum()/1000
Total_Review.head(50)

```

```python
fig, ax = plt.subplots(figsize=(80,40))
x_axis = np.arange(len(cleaned_df2["Reviews"]))  
total_review = ax.bar(x_axis[0:], Total_Review[0:], color='lightgreen')
ax.set_title('Google Play - Total Reviews For Each Category',fontsize=100)
ax.set_ylabel('Total Number of Reviews (in 1,000)',fontsize=80)
ax.set_xlabel('App Category',fontsize=80)
ax.set_xticks(x_axis+0.3)
ax.tick_params(axis="y", labelsize=50)
ax.set_xticklabels(category_key,fontsize=40)
plt.xticks(rotation=60, ha='right')
ax.grid(True)
def label(numbers,fontsize,labelheight,roundnumber):
    for x in numbers:
        height = x.get_height()
        xloc=x.get_x() + x.get_width() / 2
        label=(f'{round(height,roundnumber)}')
        ax.text(xloc,height+labelheight,label,ha='center', va='bottom',color="black",fontsize=fontsize)
label(total_review,40,100,0)
```
